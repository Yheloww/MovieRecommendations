```python
import pandas as pd
import ast

df = pd.read_csv('../datas/movies_metadata.csv')
df.columns
```

```python

# get genres in a list
df.genres = df.genres.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
df.genres = df.genres.apply(lambda x: [genre['name'] for genre in x])
# change some type
df = df.astype({'title':'string'})
# clean 
df = df.dropna(subset=['title','overview','vote_count'])
# change types
df['id'] = df['id'].astype('int64')
df['release_date'] = pd.to_datetime(df['release_date'])
# get rid of weird values
df = df.loc[(df['vote_count']) > (df['vote_count'].mean())]
# creation of a year column 
df['year'] = df['release_date'].dt.year
# deleting useless col
df = df.drop(['status','original_title','homepage','belongs_to_collection'], axis=1)

df['budget'] = df['budget'].astype('int64')
df['overview'] = df['overview'].astype('str')
```

```python
links = pd.read_csv("../datas/links.csv")

df.rename(columns = {'id':'tmdbId'}, inplace = True)
test = pd.merge(links,df, on='tmdbId',how='inner')
test.drop(['imdbId'], axis=1)
```

```python
test.to_csv('../cleaned_data/cleaned_movies.csv', index=False)

```
